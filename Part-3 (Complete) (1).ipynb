```python
# Extract the four-digit contact ID number.
import pandas as pd
import re
import json

# Raw file content
raw_data = """
contact_info
{"contact_id": 4661, "name": "Cecilia Velasco", "email": "cecilia.velasco@rodrigues.fr"}
{"contact_id": 3765, "name": "Mariana Ellis", "email": "mariana.ellis@rossi.org"}
{"contact_id": 4187, "name": "Sofie Woods", "email": "sofie.woods@riviere.com"}
{"contact_id": 4941, "name": "Jeanette Iannotti", "email": "jeanette.iannotti@yahoo.com"}
{"contact_id": 2199, "name": "Samuel Sorgatz", "email": "samuel.sorgatz@gmail.com"}
{"contact_id": 5650, "name": "Socorro Luna", "email": "socorro.luna@hotmail.com"}
"""

# Create a list of dictionaries from the raw data
data_list = [json.loads(line) for line in raw_data.strip().split('\n')[1:]]

# Create the Contacts DataFrame
contacts_df = pd.DataFrame(data_list)

# Use regex to extract the four-digit contact ID number
contacts_df['four_digit_id'] = contacts_df['contact_id'].astype(str).str.extract(r'(\d{4})')

# Split the 'name' column into 'first_name' and 'last_name'
contacts_df[['first_name', 'last_name']] = contacts_df['name'].str.split(expand=True)

# Reorder the columns for the desired output
contacts_df = contacts_df[['contact_id', 'email', 'first_name', 'last_name']]


# Print the Contacts DataFrame
print(contacts_df)
```

```python
# Convert the "contact_id" column to an int64 data type
contacts_df['contact_id'] = contacts_df['contact_id'].astype('int64')

# Display the information about the DataFrame
print(contacts_df.info(max_cols=contacts_df.shape[0]))
```

```python
# Extract the name of the contact and add it to a new column.
import re

# Assuming 'contact_info' column contains the JSON data
contact_info_df['name'] = contact_info_df['contact_info'].apply(lambda x: re.search(r'"name":\s*"([^"]*)"', x).group(1) if re.search(r'"email":\s*"([^"]*)"', x) else None)

# Print the DataFrame to verify the changes
print(contact_info_df[['name']])
```

```python
# Extract the email from the contacts and add the values to a new column.
import re

# Assuming 'contact_info' column contains the JSON data
contact_info_df['email'] = contact_info_df['contact_info'].apply(lambda x: re.search(r'"email":\s*"([^"]*)"', x).group(1) if re.search(r'"email":\s*"([^"]*)"', x) else None)

# Print the DataFrame to verify the changes
print(contact_info_df[['email']])
```

```python
import re

# Assuming 'contact_info' column contains the JSON data
contact_info_df['contact_id'] = contact_info_df['contact_info'].apply(lambda x: re.search(r'"contact_id":\s*(\d+)', x).group(1) if pd.notnull(x) and re.search(r'"contact_id":\s*(\d+)', x) else None)

# Print the modified DataFrame
print(contact_info_df[['contact_id']])
```

```python
# Create a "first"name" and "last_name" column with the first and last names from the "name" column. 
# Drop the contact_name column

# Split the 'name' column into 'first_name' and 'last_name'
contact_info_df[['first_name', 'last_name']] = contact_info_df['name'].str.split(expand=True)

# Drop the original 'name' column
contact_info_df.drop(columns=['name'], inplace=True)

# Print the modified DataFrame
print(contact_info_df)
```

```python
# Reorder the columns
contact_info_df = contact_info_df[['contact_id', 'first_name', 'last_name', 'email']]

# Print the modified DataFrame
print(contact_info_df)
```
